```python
#   Codigo que implementa esquemas numericos
#   para aproximar derivadas
#
#
#           Autor:
#   Dr. Ivan de Jesus May-Cen
#   imaycen@hotmail.com
#   Version 1.0 : 05/03/2025
#

import numpy as np
import matplotlib.pyplot as plt

# Definir funcin y su derivada analtica
def f(x):
    return np.sin(x)  # Funcin original

def df_analytical(x):
    return np.cos(x)  # Derivada correcta de f(x)

# Mtodos de diferencias finitas
def forward_diff(f, x, h):
    return (f(x + h) - f(x)) / h

def backward_diff(f, x, h):
    return (f(x) - f(x - h)) / h

def central_diff(f, x, h):
    return (f(x + h) - f(x - h)) / (2*h)

# Definir intervalo de evaluacin
a = 0.0
b = np.pi  # Corregido a pi
h = 0.1
x_vals = np.arange(a, b + h, h)  # Valores de x con paso h
df_exact = df_analytical(x_vals)

# Evitar errores en los extremos
x_forward = x_vals[:-1]  # Excluir ltimo punto para diferencias hacia adelante
x_backward = x_vals[1:]  # Excluir primer punto para diferencias hacia atrs
x_central = x_vals[1:-1]  # Excluir primer y ltimo punto para diferencias centradas

# Calcular aproximaciones numricas
df_forward = forward_diff(f, x_forward, h)
df_backward = backward_diff(f, x_backward, h)
df_central = central_diff(f, x_central, h)

# Errores absolutos
error_forward = np.abs(df_forward - df_analytical(x_forward))
error_backward = np.abs(df_backward - df_analytical(x_backward))
error_central = np.abs(df_central - df_analytical(x_central))

# Graficar derivadas
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f(x_vals), '-', label='Funcin f(x)')
plt.plot(x_vals, df_exact, 'k-', label='Derivada Analtica f\'(x)')
plt.plot(x_forward, df_forward, 'r--', label='Diferencia Hacia Adelante')
plt.plot(x_backward, df_backward, 'g-.', label='Diferencia Hacia Atrs')
plt.plot(x_central, df_central, 'b:', label='Diferencia Centrada')
plt.xlabel('x')
plt.ylabel('Derivada')
plt.legend()
plt.title("Comparacin de Mtodos de Diferenciacin Numrica")
plt.grid()
plt.savefig("diferenciacion_aproximaciones.png")
plt.show()

# Graficar errores
plt.figure(figsize=(10, 6))
plt.plot(x_forward, error_forward, 'r--', label='Error Hacia Adelante')
plt.plot(x_backward, error_backward, 'g-.', label='Error Hacia Atrs')
plt.plot(x_central, error_central, 'b:', label='Error Centrada')
plt.xlabel('x')
plt.ylabel('Error absoluto')
plt.legend()
plt.title("Errores en Diferenciacin Numrica")
plt.grid()
plt.savefig("diferenciacion_errores.png")
plt.show()
```
